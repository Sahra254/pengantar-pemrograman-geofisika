```python
import pandas as pd
from scipy.stats.mstats import gmean, hmean
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# Untuk menggunakan data yang lain misalnya saya disini menggunakan data Cs
a_mean = my_dataset.Cs.mean()
g_mean = gmean(my_dataset['Cs'])
h_mean = hmean(my_dataset['Cs'])

# 0 :.1f gunanya untuk membulatkan angka menjadi 1 digit dibelakang koma, print arithmetic mean ini gunanya untuk menampilkan nilai arithmetic mean 

print ('-------')
print ('arithmetic mean')
print ("{0:.1f} [ppm]".format(a_mean))
print ('-------')

print ('geometric mean')
print ("{0:.1f} [ppm]".format(g_mean))
print ('-------')

print ('harmonic mean')
print ("{0:.1f} [ppm]".format(h_mean))
print ('-------')

# digunakan untuk membuat area gambar
fig, ax = plt.subplots()

# disini bins auto gunanya untuk menentukan jumlah batang histogram, edge color k itu untuk warna pinggiran batangnya, measurements hits itu nama untuk legendanya nanti muncul kalau pakai ax.legend()
# alpha tadinya disini 0.8 tetapi saya ubah jadi 1.0 agar grafiknya tidak transparan
ax.hist(my_dataset.Cs, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=1.0) 

# disini untuk menentukan warna garisnya, linewidth 3 saya pakai disini agar garisnya itu tebal dan lebih jelas
ax.axvline(a_mean, color='red', label='Arithmetic mean', linewidth=3)
ax.axvline(g_mean, color='yellow',  label='Geometric mean', linewidth=3)
ax.axvline(h_mean, color='green',  label='Harmonic mean', linewidth=3)
ax.set_xlabel('Cs [ppm]')

# digunakan untuk memberi nama untuk label grafiknya
ax.set_ylabel('Kepadatan Probabilitas')

# Untuk menampilkan legenda pada grafik
ax.legend()

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# mengambil kolom cs dan menghitung varians
variance =  my_dataset['Cs'].var()

# menghitung standar deviasi
stddev =  my_dataset['Cs'].std()

# menampilkan varians dan standar deviasi, 0:.0f digunakan agar hasilnya dibulatkan tanpa angka desimal
print ('-------')
print ('Variance')
print("{0:.0f} [square ppm]".format(variance))
print ('-------')
print ('Standard Deviation')
print("{0:.0f} [ppm]".format(stddev))
print ('-------')

# untuk membuat grafiknya
fig, ax = plt.subplots()

# my dataset cs adalah data yang dipakai, bins auto artinya jumalah interval histogram ditentukan otomatis
# edge color k itu untuk warna garis tepi. hasil dari codingan ini itu grafik batang yg menunjukkan sebaran data Cs
ax.hist(my_dataset.Cs, bins='auto', density = True, edgecolor='k', label='Measurements Hist')

# mean itu rata-rata data, stddev itu standar deviasi, mean-stddev artinya batas kiri 1 sigma dibawah mean
# avxline itu garis vertikal dan purple adalah warnanya, semua codingan ini menunjukkan batas bawah distribusi.
ax.axvline(my_dataset['Cs'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=3)

# menunjukkan batas atas distribusi
ax.axvline(my_dataset['Cs'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=3)

# axvspan membuat area (blok warna), alpha 0.1 itu transparan. codingan ini menandai rentan mean
ax.axvspan(my_dataset['Cs'].mean()-stddev, my_dataset['Cs'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$')

# nama label di bagian kiri dan bawah
ax.set_xlabel('Cs [ppm]')
ax.set_ylabel('Kerapatan Probabilitas')

# untuk menampilkan legendanya
ax.legend()

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# disini kita ubah dari interpolation menjadi method krna inilah yang menyebabkan eror
# dikarenakan di versi baru numpy interpolation di hapus dan diubah menjadi method
# 25 berarti persentil ke-25(Q1), percentile disini untuk mencari nilai yang membagi data
q1 = np.percentile(my_dataset.Cs, 25, method = 'midpoint') 
q3 = np.percentile(my_dataset.Cs, 75, method = 'midpoint') 
  
iqr = q3 - q1 # Interquaritle range (IQR) 

# menampilkan hasil, 0:.0f membulatkan angka tanpa desimal, ppm itu satuan
print ('-------')
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')

# membuat grafik, fig adalah kanvas utama dan ax itu area grafiknya
fig, ax = plt.subplots()

# my dataset cs itu data yg di plot, bins auto jumlah batangnya otomatis, codingan di bawah gunanya membuat histogram
ax.hist(my_dataset.Cs, bins='auto', density=True, edgecolor='k', label='Measurements Hist') 

# axvline garis vertikal, q1 itu posisi garisnya menandai batas bawah 25% data
# q3 menandai batas atas 75& data
ax.axvline(q1, color='yellow', label='Q1', linewidth=3)
ax.axvline(q3, color='red', label='Q3', linewidth=3)

# axvspan area di antara 2 nilai, alpha=0.1 transparan, code ini menunjukaan 50% data utama berada disini
ax.axvspan(q1, q3, alpha=0.2, color='orange', label='Interquaritle range (IQR)')

# memberi nama sumbu
ax.set_xlabel('Cs [ppm]')
ax.set_ylabel('Kerapatan Probabilitas')

# menampilkan legenda (keterangan warna & garis)
ax.legend()

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# mengambil data kolom cs dan menghitung rata-rata
a_mean = my_dataset.Cs.mean()
# mengambil data kolom cs dan menghitung mediannya
median = my_dataset.Cs.median()

# hist itu tinggi tiap bin, np histogram membuat histogram dalam bentuk angka
hist, bin_edges = np.histogram(my_dataset['Cs'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2

fig, ax = plt.subplots()
ax.hist(my_dataset.Cs, bins=20, density=True, edgecolor='k', label="Measurements Hist") 
ax.axvline(modal_value, color='orange', label='Modal Value', linewidth=3)
ax.axvline(median, color='purple',  label='Median Value', linewidth=3)
ax.axvline(a_mean, color='green',  label='Arithmetic mean', linewidth=3)

# nama labelnya bisa diubah sesuka hati
ax.set_xlabel('Cs [ppm]')
ax.set_ylabel('Kerapatan Probabilitas')

# menampilkan legenda
ax.legend()

```

```python
import numpy as np
# menghitung skewness (kemencengan)
from scipy.stats import skew

# rata-rata
a_mean = my_dataset.Cs.mean()
# median
median = my_dataset.Cs.median()

# modus (estimasi dari histogram)
# hist argmax gunanya mencari bin dengan frekuensi tertinggi
hist, bin_edges = np.histogram(my_dataset['Cs'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
# mengukur sebaran data, dipakai untuk normalisasi swekness
standard_deviation = my_dataset['Cs'].std()

a1 = (a_mean - modal_value) / standard_deviation 
a2 = 3 * (a_mean - median) / standard_deviation
g1 = skew(my_dataset['Cs'])

# menampilkan hasilnya
print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

fig, ax = plt.subplots()

# mengatur tampilan outlier
my_flierprops = dict(markerfacecolor='#f8e9a1', markeredgecolor='#24305e', marker='*')
# mengatur garis median
my_medianprops = dict(color="#9526df", linewidth = 3)
# membuat box bagian utama grafik 
my_boxprops = dict(facecolor="#91ba68", edgecolor="#2f95c0")

# membuat boxplot, mengambil data cs, patch artist true gunanya supaya box bisa diwarnai
# notch true membuat lekukan di tengah box, medianprops untuk mengatur garis median
ax.boxplot(my_dataset.Cs, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops) 

# nama label bebas diganti
ax.set_ylabel('Cs [ppm]')
ax.set_xticks([1])
ax.set_xticklabels(['all Epochs'])

# menampilkan grafiknya
plt.show()
```
