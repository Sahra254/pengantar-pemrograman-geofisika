```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def my_line(x, m, q):
    y = m * x + q
    return y

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset3 = my_dataset[my_dataset.Epoch == 'three']

x = np.linspace(52.5, 62, 100)
y = my_line(x, m=0.3, q=-10.3)

fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(my_dataset3.SiO2, my_dataset3.K2O, marker='*', s=120, color="#e11a1d", edgecolor="#D1DE14DE", linewidth=1, label=r'$3^{rd}$ Epoch')
ax.plot(x, y, color='#342a77', linewidth=2)

ax.annotate(r'What is the 1$\sigma$ for this point?', xy=(47.6, 6.6), xytext=(47, 8.8), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
ax.text(52.4, 5.6, r'$ Na_2O = 0.3 \cdot SiO_2 -10.3$', size=10, rotation=33)

ax.text(0.55, 0.25, rf'$\mu_{{SiO_2}} = \frac{{a_1+a_2+\cdots+a_n}}{{n}} = {57.721:.1f}\,\mathrm{{wt.\%}}$', transform=ax.transAxes, fontsize=11)


ax.set_xlabel(r'SiO$_2$ [wt%]')
ax.set_ylabel(r'K$_2$O [wt%]')

ax.minorticks_on()
ax.grid(True, linestyle='--', alpha=0.6)


ax.legend()

plt.tight_layout()
plt.show()
```
